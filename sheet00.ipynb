```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Exercise 1
def f(x):
    return np.cos(x)*np.sin(x)
```

```python
# Exercise 2
x = np.linspace(-2*np.pi, 2*np.pi, num = 200)
y = f(x)
fig, ax = plt.subplots(1, 1)
ax.plot(x,y)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
fig.savefig('wave.png')
plt.close(fig)
```

```python
# Excersise 3
mu, sig = 5.0 , 2.0
normal_vector = np.random.normal(mu, sig, 100000)
uniform_vector = np.random.uniform(0, 10, 100000)
print(f"mean of normal_vector : {np.mean(normal_vector)} \n std. deviation of normal_vector : {np.std(normal_vector)} \n")
print(f"mean of uniform_vector : {np.mean(uniform_vector)} \n std. deviation of uniform_vector : {np.std(uniform_vector)} \n")
```
